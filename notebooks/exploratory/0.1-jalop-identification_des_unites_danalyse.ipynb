## Retirer lignes et colonnes inutiles

```python
# Retirer les sans n'ayant pas toutes les informations essentielles
data.dropna(subset=["valeur_fonciere", "code_postal", "surface_terrain", "longitude",
                    "adresse_nom_voie", "surface_reelle_bati"], inplace=True)

# retirer les colonnes devenues vides
data.dropna(axis=1, how="all", inplace=True)
# retirer les doublons
data.drop_duplicates(inplace=True)

# Corriger le type de code_postal
data.change_type("code_postal",int)

print(data.info(null_counts=True))
```

## Identification d'un identificateur unique

```python
# Nombre de valeurs uniques de chaque descripteurs par groupe de m√™me "id_parcelle"
data.groupby(by="id_parcelle").nunique()
```

```python
# Le nombre maximal de valeurs uniques de chaque descripteurs parmi tous les groupes de m√™me "id_parcelle"
data.groupby(by="id_parcelle").nunique().max()
```
